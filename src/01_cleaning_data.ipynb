```python
import pandas as pd
from utils import helper_funtions
```

```python
pd.options.display.max_columns = None  # Remove "dots" from display when printing dataframes
```

```python
PATH = helper_funtions.get_path()
```

# Read data

```python
df = pd.read_excel(PATH + 'data\Sales_outbound.xlsx')
df_validation = pd.read_excel(PATH + 'data\Sales_outbound_New.xlsx')
df.sample(n=5).head()
```

# Clear data

```python
# Homogenise column names in the data set
df = df.rename({
    'previous sales_# ':'Previous_sales',
    'Type_of_Products':'Type_products',
    'Contact_Channel':'Contact_channel',
    'Num_Contacts':'Num_contacts',
    'Satisfaction_Score':'Satisfaction_score'
    }, axis=1)

df_validation = df_validation.rename({
    'previous sales_# ':'Previous_sales',
    'Type_of_Products':'Type_products',
    'Contact_Channel':'Contact_channel',
    'Num_Contacts':'Num_contacts',
    'Satisfaction_Score':'Satisfaction_score'
    }, axis=1)
```

# Cibersecurity

```python
df.insert(
    0, 'UUID_client', 
    df['Client_ID'].apply(
    lambda x: helper_funtions.uuid_generator(x)
    )
)

columns_drop = ['Client_ID', 'Name']

for i in columns_drop:
    df.drop([i], axis=1, inplace=True)
```

```python
df_validation.insert(
    0, 'UUID_client', 
    df_validation['Client_ID'].apply(
    lambda x: helper_funtions.uuid_generator(x)
    )
)

columns_drop = ['Client_ID', 'Name']

for i in columns_drop:
    df_validation.drop([i], axis=1, inplace=True)
```

# Save

```python
df.to_csv(PATH + 'data/data_cleaning.csv', index=False, encoding='utf-8')
df_validation.to_csv(PATH + 'data/data_validation.csv', index=False, encoding='utf-8')
```
